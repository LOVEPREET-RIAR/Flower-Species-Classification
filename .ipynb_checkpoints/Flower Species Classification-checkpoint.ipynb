# Image Classification

In this Notebook, we will classify images of flowers. We will build an image classifier using tf.keras.Sequential model and load data using tf.keras.preprocessing.image.ImageDataGenerator.

### Importing Packages

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Other helping libraries
import os
import random
import glob
import shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Loading and Preapring Data

### Loading Data

```python
# This script is just for the reproducibility
if os.path.exists('/Users/lovepreet/.keras/datasets/flower_photos'):
    shutil.rmtree('/Users/lovepreet/.keras/datasets/flower_photos')
```

```python
_URL = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
zip_file = tf.keras.utils.get_file(origin=_URL, fname="flower_photos.tgz", extract=True)
```

```python
base_dir = os.path.join(os.path.dirname(zip_file), "flower_photos")
```

The dataset we downloaded contains images of 5 types of flowers:
* Rose
* Daisy
* Dandelion
* Sunflowers
* Tulips

So, let's create the labels for these 5 classes:

```python
classes = ['daisy', 'dandelion', 'roses', 'sunflowers', 'tulips']
```

Also, the dataset we have downloaded has following directory structure.

<pre style="font-size: 10.0pt; font-family: Arial; line-height: 2; letter-spacing: 1.0pt;" >
<b>flower_photos</b>
|__ <b>daisy</b>
|__ <b>dandelion</b>
|__ <b>roses</b>
|__ <b>sunflowers</b>
|__ <b>tulips</b>
</pre>

As you can see there are no folders containing training and validation data. Therefore, we will have to create our own training and validation set. Let's write some code that will do this.


The code below creates a `train` and a `val` folder each containing 5 folders (one for each type of flower). It then moves the images from the original folders to these new folders such that 80% of the images go to the training set and 20% of the images go into the validation set. In the end our directory will have the following structure:


<pre style="font-size: 10.0pt; font-family: Arial; line-height: 2; letter-spacing: 1.0pt;" >
<b>flower_photos</b>
|__ <b>daisy</b>
|__ <b>dandelion</b>
|__ <b>roses</b>
|__ <b>sunflowers</b>
|__ <b>tulips</b>
|__ <b>train</b>
    |______ <b>daisy</b>: [1.jpg, 2.jpg, 3.jpg ....]
    |______ <b>dandelion</b>: [1.jpg, 2.jpg, 3.jpg ....]
    |______ <b>roses</b>: [1.jpg, 2.jpg, 3.jpg ....]
    |______ <b>sunflowers</b>: [1.jpg, 2.jpg, 3.jpg ....]
    |______ <b>tulips</b>: [1.jpg, 2.jpg, 3.jpg ....]
 |__ <b>val</b>
    |______ <b>daisy</b>: [507.jpg, 508.jpg, 509.jpg ....]
    |______ <b>dandelion</b>: [719.jpg, 720.jpg, 721.jpg ....]
    |______ <b>roses</b>: [514.jpg, 515.jpg, 516.jpg ....]
    |______ <b>sunflowers</b>: [560.jpg, 561.jpg, 562.jpg .....]
    |______ <b>tulips</b>: [640.jpg, 641.jpg, 642.jpg ....]
</pre>

Since we don't delete the original folders, they will still be in our `flower_photos` directory, but they will be empty. The code below also prints the total number of flower images we have for each type of flower. 

```python
for cl in classes:
    img_path = os.path.join(base_dir, cl)
    images = glob.glob(img_path + "/*.jpg")
    random.shuffle(images)
    train, val = images[:round(len(images)*0.8)], images[round(len(images)*0.8):]
    
    for t in train:
        if not os.path.exists(os.path.join(base_dir, "train", cl)):
            os.makedirs(os.path.join(base_dir, "train", cl))
        shutil.move(t, os.path.join(base_dir, "train", cl))
        
    for v in val:
        if not os.path.exists(os.path.join(base_dir, "val", cl)):
            os.makedirs(os.path.join(base_dir, "val", cl))
        shutil.move(v, os.path.join(base_dir, "val", cl))
```

For convenience, let us set up the path for the training and validation sets

```python
train_dir = os.path.join(base_dir, "train")
val_dir = os.path.join(base_dir, "val")
```

### Data Augmentation

```python
BATCH_SIZE= 32
IMG_SHAPE = 150
```

Training Generator:

```python
img_gen_train = ImageDataGenerator(rescale=1./255, 
                             horizontal_flip=True, 
                             width_shift_range=.15,
                             height_shift_range=.15,
                             rotation_range=45,
                             zoom_range=0.5)

train_data_gen = img_gen_train.flow_from_directory(directory=train_dir, 
                                       batch_size=BATCH_SIZE,
                                       target_size=(IMG_SHAPE,IMG_SHAPE),
                                       shuffle=True,
                                       class_mode="sparse")
```

Let's visualize how a single image would look like 5 different times, when we pass these augmentations randomly to our dataset.

```python
def plot_images(images_arr):
    fig, axes = plt.subplots(1,5, figsize=(12,12))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
    plt.tight_layout()
    plt.show()
```

```python
augmented_images = [train_data_gen[0][0][7] for i in range(5)]
plot_images(augmented_images)
```

Validation Generator:

```python
img_gen_val = ImageDataGenerator(rescale=1./255, )

val_data_gen = img_gen_val.flow_from_directory(directory=val_dir, 
                                       batch_size=BATCH_SIZE,
                                       target_size=(IMG_SHAPE,IMG_SHAPE),
                                       class_mode="sparse")
```

## Building and Evaluating Custom-defined CNN

### Create the CNN

**Recommended ordering of layers:**

-> CONV/FC -> BatchNorm -> ReLu(or other activation) -> Dropout -> CONV/FC ->

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, 3, padding='same', kernel_initializer=tf.keras.initializers.he_normal,
                           input_shape=(IMG_SHAPE,IMG_SHAPE,3)),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    tf.keras.layers.Normalization(),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Conv2D(64, 3, padding='same'),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    tf.keras.layers.Normalization(),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Conv2D(128, 3, padding='same'),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    tf.keras.layers.Normalization(),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Flatten(),
    
    tf.keras.layers.Dense(512),
    tf.keras.layers.Normalization(),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Dense(len(classes))
])
```

```python
model.summary()
```

### Compile the Model

```python
model.compile(optimizer=tf.keras.optimizers.Adam(), 
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=["accuracy"])
```

### Callbacks

```python
early_stopping = tf.keras.callbacks.EarlyStopping(patience=5, min_delta=0.0001, restore_best_weights=True)
```

### Train the Model

```python
epochs = 100

history = model.fit(
    train_data_gen,
    epochs=epochs,
    validation_data=val_data_gen,
    callbacks=[early_stopping]
)
```

### Model Evaluation

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = history.epoch

plt.figure(figsize=(16, 10))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

# Transfer Learning 

We will use tensorflow_hub module to fetch pretrained layers for transfer learning. Here we will use **MobileNet**, which is a type of convolutional neural network designed for mobile and embedded vision applications.

First, we need to reformat all images to the resolution expected by MobileNet (224, 224).

```python
BATCH_SIZE = 32
IMAGE_RES = 224

# Training Generator
img_gen_train = ImageDataGenerator(rescale=1./255, 
                             horizontal_flip=True, 
                             width_shift_range=.15,
                             height_shift_range=.15,
                             rotation_range=45,
                             zoom_range=0.5)

train_data_gen = img_gen_train.flow_from_directory(directory=train_dir, 
                                       batch_size=BATCH_SIZE,
                                       target_size=(IMAGE_RES,IMAGE_RES),
                                       shuffle=True,
                                       class_mode="sparse")

# Validation Generator
img_gen_val = ImageDataGenerator(rescale=1./255, )

val_data_gen = img_gen_val.flow_from_directory(directory=val_dir, 
                                       batch_size=BATCH_SIZE,
                                       target_size=(IMAGE_RES,IMAGE_RES),
                                       class_mode="sparse")
```

We will use tensorflow_hub module to fetch pretrained layers for transfer learning. Here we will use **MobileNet**, which is a type of convolutional neural network designed for mobile and embedded vision applications. 

```python
import tensorflow_hub as hub
```

```python
URL = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"

mob_net_layers = hub.KerasLayer(URL, input_shape=(224, 224, 3))
mob_net_layers.trainable = False
```

### Create the CNN

Next, we will attach a classification head.

```python
model = tf.keras.models.Sequential([
    mob_net_layers,
    tf.keras.layers.Dense(len(classes))
])
```

```python
model.summary()
```

### Compile the Model

```python
model.compile(optimizer=tf.keras.optimizers.Adam(), 
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=["accuracy"])
```

### Callbacks

```python
early_stopping = tf.keras.callbacks.EarlyStopping(patience=5, min_delta=0.0001, restore_best_weights=True)
```

### Train the Model

```python
epochs = 100

history = model.fit(
    train_data_gen,
    epochs=epochs,
    validation_data=val_data_gen,
    callbacks=[early_stopping]
)
```

### Model Evaluation

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = history.epoch

plt.figure(figsize=(16, 10))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
